```python
import json
import psycopg2
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
import numpy as np
from urllib.request import urlopen
from urllib.request import HTTPError
from bs4 import BeautifulSoup
import time
import matplotlib.pyplot as plt
```

```python
kha = ["병원"]

for page in range(1, 34):
    while (True):
        try:
            html = urlopen("http://www.kha.or.kr:8080/info/sub_01.asp?page=" + str(page))
        except HTTPError as e:
            print(e)
            continue
        break
    bsObject = BeautifulSoup(html, "html.parser")

    for element in bsObject.find_all("font", {"style":"font-height:18px; font-size:12px;"}):
        kha.append(element.get_text())
        print(element.get_text())
```

```python
list(set(kha))
kha_df = pd.DataFrame(kha, columns=['Hospital'])
print(kha_df.shape[0])
kha_df = kha_df.drop_duplicates()
kha_df = kha_df.reset_index(drop = True)
kha_df.head(10)
```

```python
dbconfig = json.loads(open("res/dbconfig.properties").read())['cssc_jaunt']
dbconfig.pop("drivername")
dbconfig['user'] = dbconfig['username']
dbconfig.pop("username")
connection = psycopg2.connect(**dbconfig)
connection
```

```python
article_df = pd.read_sql(sql = "SELECT \"title\", \"content\", \"aDate\" FROM article_entries WHERE \"aDate\" BETWEEN '2015-05-20' AND '2015-12-31 23:59:59'", con = connection)
```

```python
def get_tag(text, ntags):
    spliter = Okt()
    nouns = spliter.nouns(text)
    
    count = Counter(nouns)
    return_list = []
    
    for n, c in count.most_common(ntags):
        if (n == "환자"):
            temp = {'tag': n, 'count': c}
            return_list.append(temp)

    return return_list
```

```python
import time
start_time = time.time()

noun_list = []
count_list = []

for i in range(5, 13):
    article_df_month = article_df.loc[(article_df['aDate'].dt.month == i)]

    noun_count = 200
    tags = get_tag(article_df_month.to_string(), noun_count)
    #output_file_name = "article_count.txt"
    #open_output_file = open(output_file_name, 'w', -1, "utf-8")

    for tag in tags:
        noun = tag['tag']
        count = tag['count']
        print(str(i) + "월", noun, ":", count)
        noun_list.append(str(i))
        count_list.append(count)
        #open_output_file.write('{} {}\n'.format(noun, count))    
    #open_output_file.close() 
    #print(time.time() - start_time)
```

```python
patient_df = pd.DataFrame({"Month" : noun_list, "Patient" : count_list})
patient_df
```

```python
total = patient_df.sum(axis = 0)[1]
day = 12 + 30 + 31 + 31 + 30 + 31 + 30 + 31
month_avr = (total / day) * 30
print("Month Average :", month_avr)

select_month = patient_df[patient_df['Patient'] > month_avr]
select_month
```
